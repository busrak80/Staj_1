```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75] #yas listesi olusturuldu
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90] #kilo listesi oluşturuldu
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi =np.array(kiloListesi)
#numpy dizilerine dönüştürüldü
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") #plot ile grafiği çizdirebiliriz. "g"ise green yani yeşil rengi temsil eder
#x ekseni yaş listesi y ekseni kilo listesi
plt.xlabel("Yas") #xekseni etiketi
plt.ylabel("Kilo") #yekseni etiketi
plt.title("Kilonun Yaşa Göre Değişim Grafiği")#Grafiğe başlık verilebilir
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2 
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g") #"g*-" ile kesikli çizgi grafiğide oluşturulabilir
plt.xlabel("numpyDizisi1")
plt.ylabel("numpyDizisi2")
```

```python
#subplot ile alt grafik çizdirilebilir
plt.subplot(1,2,1) #1 sıra 2 kolon 1.grafik çizimi
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) #1 sıra 2 kolon 2.grafik çizimi
plt.plot(numpyDizisi2,numpyDizisi1,"b--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9 ]) #width ve height alanları
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

# İç içe grafikler

```python
figure2 = plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.3,0.3,0.3,0.3])
eksen1.plot(numpyDizisi1, numpyDizisi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana grafik başlık")
eksen2.plot(numpyDizisi2, numpyDizisi1,"g")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("alt grafik başlık")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("x ekseni")

plt.tight_layout()
    
```

```python
yeniFigur = plt.figure()#(figsize =(3,3))
#grafik boyutu ayarlanabilir
#dpi ile çözünürlük ayarlanabilir

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 **2,label ="numpyDizisi **2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 **3,label ="numpyDizisi **3")
yeniEksen.legend(loc=6)
#legend ile hangiçizgi hangi numpy dizisine ait olduğunu belirtmek için
```

```python
yeniFigur.savefig("benimfigur.png", dpi=200) #kaydetmek için savefig kullanılır
```
