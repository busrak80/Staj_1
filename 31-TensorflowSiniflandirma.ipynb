```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
#model nesnesinin eğitim sırasında kaydettiği kayıp (loss) değerlerini içeren bir pandas DataFrame oluşturmayı amaçlar
```

```python
modelKaybi.plot()
```

```python
model = Sequential()
#Sequential modeli oluşturulur.

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))
#İlk katman olarak Dense katmanı eklenir. 
#Bu katman 30 birimden oluşur ve "relu" aktivasyon fonksiyonu kullanır. Ayrıca, Dropout katmanı ile aşırı uyum riski azaltılır 

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
#Kayıp değerlerini içeren bir DataFrame oluşturma
```

```python
kayipDf.plot()
```

```python
tahminlerimiz = model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
# rapor, doğruluk (accuracy), hassasiyet (precision), geri çağırma (recall),
#F1 skoru ve destek (support) gibi önemli sınıflandırma metriklerini içerecektir
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
#Karmaşıklık matrisi, modelin her sınıfı 
#ne kadar doğru sınıflandırdığını ve hangi sınıfları karıştırdığını gösteren önemli bir matristir
```
