```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
dataFrame = pd.read_csv("merc.csv") #veriyi oku ve dataframe değiskenine aktar
```

```python
dataFrame.head() # dataframedeki ilk bes satiri listele
```

```python
dataFrame.describe() #istatiksel özet
```

```python
dataFrame.isnull().sum() #13119 tane veriyi kontrol etmek mümkün olmadığı için .sum() ile toplam null verileri görüntüleyebiliriz
```

```python
plt.figure(figsize=(7,5)) #grafik boyutu
sbn.histplot(dataFrame["price"]) #price histogram olarak inceleme
```
